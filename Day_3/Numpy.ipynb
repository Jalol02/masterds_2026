#### Ex 3

```python
arr = np.random.randint(100, 201, 20)
```

```python
even_nums = arr[arr % 2 == 0]
print("Even numbers:", even_nums)
```

```python
sliced = arr[2:10]
print("Elements 3rd to 10th:", sliced)
```

```python
dataset = np.genfromtxt('../Datasets/data.csv', delimiter=';') 
```

```python
dataset
```

```python
print(f"Sum: {np.sum(dataset)}")
print(f"Mean: {np.mean(dataset)}")
print(f"Median: {np.median(dataset)}")
```

```python
col_data = dataset[:, 0]  # the first column
print(f"\nColumn sum: {np.sum(col_data)}")
print(f"Column mean: {np.mean(col_data)}")
print(f"Column median: {np.median(col_data)}")
```
