#### Task 3

```python
def on_button_click():
    welcome_label.config(text = "Button clicked")
```

```python
root = tk.Tk()
root.title("Button Example")
click_me_button = tk.Button(root, text="Click Me", command=on_button_click)
click_me_button.pack()
root.mainloop()
```
