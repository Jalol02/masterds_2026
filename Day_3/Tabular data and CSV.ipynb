#### Ex 3

```python
numbers = list(range(1, 1001))
```

```python
with open('numbers.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Number', 'Square'])  
    for num in numbers:
        writer.writerow([num, num**2])  

print("numbers.csv created successfully!")
```
