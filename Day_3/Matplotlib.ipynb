```python
import matplotlib.pyplot as plt
```

```python
# Step 2: Create two lists (x and y)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Step 3: Use the plt.plot() function to plot x against y
plt.plot(x, y)
# Step 4: Customize your plot with a title and axis labels
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Step 5: Display the plot.
plt.show() 
```

```python
# Step 1: Categorical data and values
categories = ['Cat.A', 'Cat.B', 'Cat.C', 'Cat.D']
values = [20, 35, 30, 15]
# Step 2: Create a bar chart
plt.bar(categories, values)
# Step 3: Enhance the chart
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(rotation=45) # Rotate for better readabilit
# Step 4: Show the plot
plt.show()
```
