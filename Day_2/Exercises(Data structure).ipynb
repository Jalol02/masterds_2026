```python
def insert_at(l, position, e): # 1st method to insert into list
    left_part = l[:position]
    left_part.append(e)
    right_part = l[position:]
    return left_part + right_part
```

```python
def insert_at_2(l, position, e): # 2nd method to insert into list
    return l[:position] + [e] + l[position:] 
```

```python
my_list.insert(2,1090) # 3rd method to insert into list
my_list
```

```python
def subtrack(a, b):
    return a-b

subtrack(b=10,a=7)
```
